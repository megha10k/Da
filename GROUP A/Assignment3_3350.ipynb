# 1. Provide summary statistics

```python
#This code groups the DataFrame df by the 'Commodity' column and computes several aggregate statistics for the 'min_price' column 
#using the .groupby() and .agg() methods.
df_min_price = df.groupby(['Commodity'], as_index=False).agg(mean=('min_price', 'mean'),minimum=('min_price', 'min'), maximum=('min_price', 'max'),median=('min_price', 'median'),standard_deviation=('min_price', 'std'))
```

```python
# Statistical data for min_price attribute over Commodity
df_min_price.head(20)
```

```python
df_min_price
```

```python
# plt.plot(df_min_price['mean'])
# plt.axhline(df_min_price['median'])
df_min_price.plot('Commodity',['mean','median','standard_deviation'],figsize=(15, 8))
```

```python
#This code groups the DataFrame df by the 'APMC' column and computes several aggregate statistics for the 'min_price' column using the 
#.groupby() and .agg() methods.
data= df.groupby(['APMC'], as_index=False).agg(mean=('min_price', 'mean'),minimum=('min_price', 'min'), maximum=('min_price', 'max'),median=('min_price', 'median'),standard_deviation=('min_price', 'std'))
```

```python
data
```

```python
#This code creates a line plot using the .plot() method of the DataFrame data. The x-axis of the plot is the 'APMC' column, and 
#the y-axis is the values of the 'mean', 'median', and 'standard_deviation' columns.
data.plot('APMC',['mean','median','standard_deviation'],figsize=(15, 8))
```

# 2. Iris dataset

```python
# Loading Iris Dataset
iris = pd.read_csv('iris.csv')
```

```python
iris
```

```python
# This code groups the DataFrame iris by the 'Species' column and computes the mean (average) of each column for 
# each group using the .groupby() and .agg() methods.
mean_df = iris.groupby(['Species'], as_index=False).agg('mean')
```

```python
mean_df 
```

```python
mean_df[['SepalLengthCm','SepalWidthCm','PetalWidthCm','PetalLengthCm']].plot(figsize=(12,7))
```

```python
std_df = iris.groupby(['Species'], as_index=False).agg('std')
```

```python
std_df
```

```python
#This code creates a line plot using the .plot() method of the DataFrame std_df. The x-axis of the plot is the 'Species' column, 
#and the y-axis is the values of the 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', and 'PetalWidthCm' columns.
std_df.plot('Species',['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],figsize=(15, 8))
```

```python
med_iris = iris.groupby(['Species'], as_index=False).agg('median')
```

```python
med_iris
```

```python
med_iris.plot('Species',['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],figsize=(15, 8))
```
