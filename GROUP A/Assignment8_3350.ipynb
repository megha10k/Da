```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

-> pandas is a data analysis library used to manipulate and analyze data in Python. It provides data structures such as DataFrame and Series for handling tabular data.

-> seaborn is a data visualization library based on matplotlib. It provides high-level interface for creating statistical graphics such as scatter plots, heatmaps, and bar plots.

-> matplotlib is a plotting library used to create static, animated, and interactive visualizations in Python. It provides a range of visualization tools for creating line plots, scatter plots, histograms, and more.

-> numpy is a numerical computing library used for scientific computing in Python. It provides support for large, multi-dimensional arrays and matrices, and a range of mathematical functions for working with these arrays.

Loading "titanic" dataset

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df.columns
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df_c= df.groupby(by='Survived').agg(count=('PassengerId','count'))
# The code df.groupby(by='Survived').agg(count=('PassengerId','count')) groups the data in a DataFrame df by the 'Survived' column, 
# and calculates the count of rows in each group using the 'PassengerId' column. The result is a new DataFrame with a single column 
# 'count' and two rows, one for each possible value of 'Survived' (0 and 1).
df_c
```

```python
sns.distplot(df['Age'])
# plot of the distribution of the 'Age' variable in the given dataset 'df', using the seaborn library. 
# It shows the frequency distribution of the 'Age' variable, with the x-axis representing the range of ages 
# and the y-axis representing the frequency of occurrence of those ages. 
# The plot can give an idea of the central tendency, spread, and shape of the distribution of ages in the dataset.
```

```python
sns.countplot(y="Survived", data=df)
```

```python
sns.catplot(x ="Survived",col='Pclass',
kind ="count", data = df)
# This plot shows how many passengers survived and died for each class in the dataset, 
# giving us an idea about the survival rate among different classes.
```

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
```

```python
sns.catplot(x ="Sex", hue ="Survived",
kind ="count", data = df)
```

```python
sns.countplot(x="Survived", hue="Sex", data=df)
```

```python
df['isAdult'] = np.where(df['Age']>=18,1,0)
# creates a new column named 'isAdult' in the DataFrame 'df' based on a condition.
# If the 'Age' of a person is greater than or equal to 18, the value of 'isAdult' for that 
# person is set to 1, otherwise it is set to 0.
```

```python
sns.catplot(x ="isAdult", hue ="Survived",
kind ="count", data = df)
```

```python
sns.catplot(x ="Sex",col='isAdult',
kind ="count", data = df)
```

```python
sns.catplot(x ="Survived",col='Embarked', hue='Sex',
kind ="count", data = df)
# this plot shows the count of male and female passengers who survived and did not survive for each Embarked port.
# It helps to visualize if there is any relationship between Embarked port, Sex, and the survival rate.
```

```python
sns.distplot(df['Fare'])
# The values on the x-axis of the plot represent the fare paid by passengers, while the values on the y-axis represent the density of passengers for each fare value. 
# The plot shows the distribution of fare values in the dataset, indicating how many passengers paid a particular fare amount. 
# The peaks in the plot indicate the fare values that were more common among the passengers.
```

```python
df.corr()
```

```python
sns.jointplot(x = "Pclass", y = "Fare",
              kind = "scatter", data = df)

# This plot shows the relationship between the two variables 'Pclass' and 'Fare'. The x-axis represents the values of 'Pclass', while the y-axis represents the values of 'Fare'. 
# Each point on the plot represents a passenger in the dataset. The joint plot displays both a scatter plot and a histogram of the distribution of each variable. 
# The scatter plot shows the individual data points while the histograms show the distribution of each variable along the corresponding axis. 
# This plot helps to visualize the relationship between the passenger class and fare they paid.
```

```python
sns.jointplot(x = "Embarked", y = "Fare",
              kind = "scatter", data = df)
```

```python
sns.jointplot(x = "Age", y = "Fare",
              kind = "scatter", data = df)
```
