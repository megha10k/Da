```python
# Importing Pandas and other Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("boston.csv")
df
```

The prices of the house indicated by the variable MEDV is our target variable and the remaining are the feature variables based on which we will predict the value of a house.


```python
df.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
df.drop(['a'],axis=1, inplace=True)
```

```python
#This code creates a distribution plot of the 'medv' column in the 
# DataFrame df using the distplot() function from the Seaborn library.
#The first line sets the figure size for the plot using the set() method of the Seaborn library.
#The second line creates the distribution plot using the distplot() function. The 'medv' column is passed as the data to plot, and the bins parameter is set to 20 to divide the range of the data into 20 equally sized bins. The resulting plot shows the distribution of the 'medv' values as a histogram.
sns.set(rc={'figure.figsize':(12,9)})
sns.distplot(df['medv'], bins=20)
plt.show()
```

To fit a linear regression model, we select those features which have a high correlation with our target variable MEDV. By looking at the correlation matrix we can see that RM has a strong positive correlation with MEDV (0.7) where as LSTAT has a high negative correlation with MEDV(-0.74).

```python
plt.figure(figsize=(15, 10))
plt.title("rm")
plt.xlabel("rm")
plt.ylabel('medv')
plt.scatter(df['rm'],df['medv'])
```

```python
# This code creates a scatter plot using the scatter() function from the Matplotlib library to plot the 'medv' values 
# against the 'lstat' values in the DataFrame df.
plt.figure(figsize=(15, 10))
plt.title("lstat")
plt.xlabel("lstat")
plt.ylabel('medv')
plt.scatter(df['lstat'],df['medv'])
```

```python
rm= df['rm']
medv= df['medv']
```

```python
x_train, x_test, y_train, y_test = train_test_split(rm, medv, test_size = 0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
# Converting 2D Array
#This code reshapes the NumPy arrays x_train, y_train, x_test, and y_test from one-dimensional arrays to two-dimensional arrays 
#with one column and a number of rows equal to the length of the original arrays.
#The .reshape() method is called on each array, and the argument -1 is passed as the first dimension, 
#which means that the number of rows is inferred from the length of the original array. The second dimension is set to 1 to create an array with one column.
x_train = np.array(x_train).reshape(-1, 1) 
y_train = np.array(y_train).reshape(-1, 1) 
x_test =  np.array(x_test).reshape(-1, 1) 
y_test =  np.array(y_test).reshape(-1, 1) 
```

```python
#These two lines of code fit a linear regression model to the training data.
# Linear regression is a type of statistical analysis that is used to model the relationship between a dependent variable and 
#one or more independent variables. 
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

Using rm as independent variable for medv

```python
plt.title("predicted results")
plt.xlabel("rm")
plt.ylabel('medv')
plt.scatter(x_train, y_train, color='blue')
plt.scatter(x_test, y_predict, color='red',label="aa")
plt.scatter(x_test, y_test, color="yellow")
```

--> Using lstat as independent variable for medv

```python
# This code is performing a simple linear regression using the "lstat" variable as the independent variable and "medv" as the dependent variable.
lstat= df['lstat']
x_train, x_test, y_train, y_test = train_test_split(lstat, medv, test_size = 0.2)
x_train = np.array(x_train).reshape(-1, 1) 
y_train = np.array(y_train).reshape(-1, 1) 
x_test =  np.array(x_test).reshape(-1, 1) 
y_test =  np.array(y_test).reshape(-1, 1)
model = LinearRegression()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
plt.title("predicted results")
plt.xlabel("lstat")
plt.ylabel('medv')
plt.scatter(x_train, y_train, color='blue')
plt.scatter(x_test, y_predict, color='red')
plt.scatter(x_test, y_test, color="yellow")
```

--> The model performance for testing set

```python
X = pd.DataFrame(np.c_[df['lstat'], df['rm']], columns = ['lstat','rm'])
x_train, x_test, y_train, y_test = train_test_split(X, medv, test_size = 0.2)
model = LinearRegression()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
```

```python
rmse = (np.sqrt(mean_squared_error(y_test, y_predict)))

print('Root Mean Squared Error is {}'.format(rmse))
```
